```python
import librosa

y, sr = librosa.load( librosa.util.example_audio_file() )
#y, sr = librosa.load("data/sound/sine1000.wav")
#y, sr = librosa.load("data/sound/pack/jalastram/lo/362371__jalastram__fx369.wav")
```

```python
D = librosa.stft(y)
```

```python
D.shape
```

```python
D.reshape(1025*130).shape
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import librosa.display

%matplotlib notebook

librosa.display.specshow(librosa.logamplitude(np.abs(D)**2,
                                               ref_power=np.max),
                          y_axis='log', x_axis='time')
plt.title('Power spectrogram')
plt.colorbar(format='%+2.0f dB')
plt.tight_layout()
```

```python
import math

FREQUENCY = 1000

FRECUENCY_BIN = math.ceil ( FREQUENCY * D.shape[0] / 11025 )
TIME = 40

frecuency = FRECUENCY_BIN * 11025 / D.shape[0]
tiempo = TIME * librosa.core.get_duration(y) / D.shape[1] 
print(frecuency, "hz" ":", tiempo, "s", "es", librosa.logamplitude(np.abs(D[FRECUENCY_BIN, TIME])**2), "db" ) 
```
